```python
import re
```

```python
# extract all word

def check_contain_num(text):
    return bool(re.search(r'(^|[^\w])\d+(?:(\.|\,)\d+)?\b', text))

def extract_nutri_name(text):
    text = text.lower()
    toks = re.findall(r'(?:^|\,)([\w ]+)(?:$|\,)', text)
    return [tok.strip() for tok in toks if not check_contain_num(tok)]
```

```python
check_contain_num('4 mg')
```

```python
extract_nutri_name('energy,811 kcal,axit lauric,41.421 g,total fat,90.14 g,saturated fat,80.138 g',)
```

```python
all_nutri_name = []

for nutri_text in nutri_list:
    all_nutri_name += extract_nutri_name(nutri_text)
```

```python
from collections import Counter
```

```python
nutri_stat = Counter(all_nutri_name)
```

```python
# sort dict by value
nutri_stat = {k: v for k, v in sorted(nutri_stat.items(), key=lambda item: item[1], reverse=True)}
```

```python
nutri_stat
```

```python
# write to file as pandas dataframe

nutri_stat_df = pd.DataFrame.from_dict(nutri_stat, orient='index', columns=['count'])
nutri_stat_df.to_csv('data/nutri_stat.csv')
```

```python
element_df = pd.read_html('https://www.vedantu.com/chemistry/118-elements-and-their-symbols-and-atomic-numbers')
```

```python
element_df = element_df[0]
```

```python
ele_data = element_df.to_dict('records')
```

```python
ele_data[1]
```

```python
ele_dict = {}
for item in ele_data:
    ele_dict[item[2].lower()] = item[0].lower()
```

```python
ele_dict
```

```python
clean_nutri_name = []

for nutri_name in all_nutri_name:
    if nutri_name in ele_dict:
        clean_nutri_name.append(ele_dict[nutri_name])
    else:
        clean_nutri_name.append(nutri_name.lower())
```

```python
clean_nutri_name = list(set(clean_nutri_name))
```

```python
clean_nutri_name = [x for x in clean_nutri_name if x != '']
```

```python
clean_nutri_name
```

```python
# revert dict
ele_dict = {v: k for k, v in ele_dict.items()}
```

```python
ele_dict
```

```python
list_add = list()
for nutri_name in clean_nutri_name:
    if nutri_name in ele_dict:
        list_add.append(nutri_name + " ( " + ele_dict[nutri_name] + " )")
    else:
        list_add.append(nutri_name)
```

```python
# write list to csv pandas

nutri_list_df = pd.DataFrame(list_add, columns=['nutri_name'])
nutri_list_df.to_csv('data/nutri_list_raw.csv', index=False)
```
