```python
import re
def clear_text(text):
    # text = text.replace(" ", "")
    try:
        text = text.replace("\n", " ")
        text = text.replace("\t", " ")
        text = text.replace("\r", " ")
        # remove multiple space
        text = re.sub(' +', ' ', text)
        return text
    except:
        return text
```

```python
# search each term in raw_dict_terms

def get_match_term_query(raw_term, rank_list_size=10):
    j = {
        # "explain": True,
        "query": {
            "bool": {
                "should": [
                    {"match": {"full_name_no_unique": {"query": raw_term, "boost": 1}}},
                ],
                # "minimum_should_match": 1,
            }
        },
        "_source": {"includes": ["full_name_no_unique"]},
        'from': 0,
        'size': rank_list_size
    }

    return j
```

```python
def search_term(raw_term, rank_list_size=10):
    j = get_match_term_query(raw_term, rank_list_size)
    res = es.search(index=index_name, body=j)
    return res
```

```python
def clean_sep_string_head_tail(text):
    return text.replace(sep_string, " , ").strip(" , ")
```

```python
def extract_name_and_score(res):
    return [(clean_sep_string_head_tail(i['_source']['full_name_no_unique']), i['_score']) for i in res['hits']['hits']]
```

```python
clean_pair_list = [extract_name_and_score(i) for i in ret_list]
```

```python
# pair the query and result
query_result_pair = list(zip(raw_dict_terms, clean_pair_list))
```

```python
# extract match pair
match_pair_list = list()
for query, results in query_result_pair:
    for result in results[:3]:
        query = (clean_sep_string_head_tail(query))
        if query == result[0]:
            continue
        else:
            match_pair_list.append({
                "query": query, 
                "match": result[0],
                 "score": result[1]})
```
