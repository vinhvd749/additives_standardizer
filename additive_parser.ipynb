```python
# search each term in raw_dict_terms

def get_match_term_query(raw_term, rank_list_size=10):
    j = {
        # "explain": True,
        "query": {
            "bool": {
                "should": [
                    # {"match": {"full_name_no_unique": {"query": raw_term, "boost": 1}}},
                    {"match": {"vi_name": {"query": raw_term, "boost": 1}}},
                    {"match": {"en_name": {"query": raw_term, "boost": 1}}},
                    {"match": {"e_number": {"query": raw_term, "boost": 3}}},
                ],
                "minimum_should_match": 1
            }
        },
        "_source": {"includes": ["vi_name", "en_name",  "e_number"]},
        'from': 0,
        'size': rank_list_size
    }

    return j
```

```python
sep_string = ' dq32eqw4 '
import re

def preprocess_search_text(text):
    # subtitude \w+token with sep_string
    text = re.sub(r'\w+token', sep_string, text)
    text = re.sub(r' \| ', sep_string, text)
    # remove multiple sep_string
    text = re.sub(r'({})+'.format(sep_string), sep_string, text)
    return sep_string + text + sep_string
```

```python
def check_string_issubset(s1, s2):
    toks1 = re.findall(r'[\w\d]+', s1)
    toks2 = re.findall(r'[\w\d]+', s2)
    return set(toks1).issubset(set(toks2))
```

```python
def search_term(raw_term, rank_list_size=20):
    raw_term = preprocess_search_text(raw_term)
    j = get_match_term_query(raw_term, rank_list_size)
    res = es.search(index=index_name, body=j)

    ret_list = []

    for i in res['hits']['hits']:
        i['_source']['vi_name'] = i['_source']['vi_name'].replace(sep_string, " ")
        i['_source']['en_name'] = i['_source']['en_name'].replace(sep_string, " ")
        i['_source']['e_number'] = i['_source']['e_number'].replace(sep_string, " ")

        d = {
                'score': i['_score'],
                'vi_name': i['_source']['vi_name'].replace(sep_string, " ").strip(),
                'en_name': i['_source']['en_name'].replace(sep_string, " ").strip(),
                'e_number': i['_source']['e_number'].replace(sep_string, " ").strip(),
        }

        if d['score'] >= 10 and (check_string_issubset( d['vi_name'], raw_term) or
                                  check_string_issubset( d['en_name'], raw_term) or
                                    check_string_issubset( d['e_number'], raw_term)):
            del d['score']
            ret_list.append(d)
    return ret_list
```

```python
def find_all_additives(text):
    text = tokenizer(text)
    additives = search_term(text, rank_list_size=20)

    for additive in additives:
        text = text.replace(additive['vi_name'], '')
        text = text.replace(additive['en_name'], '')
        text = text.replace(additive['e_number'], '')

    text = text.strip()
    text = re.sub(r'\w+token', ' | ', text)
    # text = re.sub(r' \| ', sep_string, text)
    # remove multiple sep_string
    toks = re.split(r'\|', text)
    toks = [i.strip() for i in toks if i.strip() != '']
    d = {
        'additives': additives,
        'other_ingredients': toks
    }
    return d
    
```
