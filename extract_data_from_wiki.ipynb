```python
# write data from wiki to dataframes

import pandas as pd


df = pd.read_html('https://en.wikipedia.org/wiki/International_Numbering_System_for_Food_Additives')


```

```python
df[1]
```

```python
e_num_df_list = pd.read_html('https://en.wikipedia.org/wiki/E_number')
```

```python
e_num_df_list[1]
```

```python
count = 0
for tb in e_num_df_list[1:-4]:
    # write to csv
    tb.to_csv('data/e_number/' + str(count) + '.csv', index=False, header=False)
    count += 1
```
